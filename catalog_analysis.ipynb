```python
import os
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
from fermipy.gtanalysis import GTAnalysis
from fermipy.plotting import ROIPlotter, SEDPlotter
from IPython.display import Image
```

```python
gta = GTAnalysis('config.yaml')
```

```python
gta.setup()
```

```python
gta.print_roi()
```

## Optimize and Fit

```python
gta.optimize()
```

```python
gta.print_model()
```

```python
gta.free_sources(free=False)
gta.free_sources(distance=1.0,pars='norm')
gta.free_sources(distance=1.0,pars='shape')
gta.fit(optimizer='NEWMINUIT')
```

```python
tsmap_postfit = gta.tsmap('uks_postfit',model={'SpatialModel' : 'PointSource', 'Index' : 2.0})
```

```python
plt.clf()
fig = plt.figure(figsize=(14,6))
ROIPlotter(tsmap_postfit['sqrt_ts'],roi=gta.roi).plot(levels=[0,3,5,7],vmin=0,vmax=5,subplot=121,cmap='magma')
plt.gca().set_title('Sqrt(TS)')
ROIPlotter(tsmap_postfit['npred'],roi=gta.roi).plot(vmin=0,vmax=100,subplot=122,cmap='magma')
plt.gca().set_title('NPred')
plt.show()
```
